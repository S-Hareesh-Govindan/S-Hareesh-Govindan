```python
import numpy as np
import pandas as pd                                 #DATA HANDLING
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans                  #CLUSTER DIAGRAM
from sklearn.metrics import silhouette_score        #FOR SILHOUETTE DIAGRAM
```

Data Collections & Analysis

```python
# Loading the data from csv file to a pandas DataFrame
c_ds = pd.read_csv('/content/customer_ds.csv')
```

# Choosing the Annual income column & Spending Score column

```python
X= c_ds.iloc[:,[3,4]].values
```

```python
print(X)
```

X axis : Annual Income

Y axis : Spending Score

Choosing The number of clusters

WCSS -> Within Clusters Sum of Squares

```python
# finding wcss value for different number of clusters
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)  # Set n_init explicitly
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# to plot an elbow graph to find the minimum value
sns.set()
plt.plot(range(1,11),wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimum Number of Clusters = 5

Training the k-Means Clustering Model
An UNSUPERVISED LEARNING MODEL


```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0, n_init=10)  # Set n_init explicitly
Y = kmeans.fit_predict(X)
print(Y)
```

5 CLUSTERS - 0,1,2,3,4
Visualizing all clusters

```python
# plotting all the clusters and their Centroids
plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='blue', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='brown', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='red', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='yellow', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='black', label='Cluster 5')

# plot the centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
